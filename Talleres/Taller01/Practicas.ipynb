##Ejercicio 2 Dado el direccionario de
datos = {'a':10,'b':25,'c':7,'d':18}

###a)Crea una lista con las claves cuyo valor sea mayor a 12
###b)Crea un nuevo diccionario con esas claves y sus valores al cuadrado



```python
#a
datos = {'a':10,'b':25,'c':7,'d':18}
value_key = []
for clave,valor in datos.items():
  if valor > 12:
    value_key += [clave]
print(value_key)
#b
new_dicc = {}
for clave,valor in datos.items():
  if clave in value_key:
    new_dicc[clave] = valor
    new_dicc[clave] = valor**2
print(new_dicc)

```

##Practica - Modulo 3- Indexing, Slicing, Reshape

###Ejercicio 1: Dado A = np.arange(1, 26).reshape(5, 5)
###a) Extrae la sub-matriz de las 3 últimas filas y las 3 primeras columnas.
###b) Extrae la segunda columna completa.
###c) Extrae los elementos de los ángulos (esquinas) de la matriz

```python
import numpy as np
A = np.arange(1,26).reshape(5,5)
print(A)
#a slicing
sub_matrix = A[2::,::2].copy()
print(sub_matrix)
#b fancy indexing
second_column = A[::,1].copy()
print(second_column)
#c fancy indexing
corners_matrix = A[[0,4],:][:,[0,4]].copy()
print(corners_matrix)
```

###Ejercicio 2: Con la misma matriz A:
###a) Crea una máscara booleana para los valores mayores a 12.
###b) Reemplaza esos valores por -1 en una copia del array.
###c) Usa np.where para reemplazar impares por 0 y mantener los pares

```python
A = np.arange(1,26).reshape(5,5)
#a mascara booleana
mask = A > 12
print(mask)
values = A[mask]
print(values)
#b modify values as -1 in an copy mask
values_copy = A.copy()
values_copy[values_copy > 12] = -1
print(values_copy)
#c np.where to replace odds as 0 and keep evens
new_array = np.where(A%2 != 0,0,A)
print(new_array)
```

###Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).
###a) Calcula la media, máximo y mínimo de cada columna.
###b) Calcula la suma acumulada de la primera fila.
##c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.Fórmula: (x - min_col) / (max_col - min_col)

```python
import numpy as np
np.random.seed(7)

M5x4= np.random.randint(1,100,(5,4))
print(M5x4)
#a
#media
media = M5x4.mean(axis=0)
print("Media: ",media)
#max
max = M5x4.max(axis=0)
print("Max: ",max)
#min
min = M5x4.min(axis=0)
print("Min: ",min)
#b suma acumulada de primera fila
suma_acumulada = np.cumsum(M5x4[0])
print("Suma acumulada de la primera fila: ", suma_acumulada)
#c Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
normalized_M5x4 = (M5x4 - min) / (max - min)
print("Matriz normalizada:")
print(normalized_M5x4)
```
